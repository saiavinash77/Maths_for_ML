# addition of two vectors

```python
vector_1 = np.array([0,0,2,3])

vector_2 = np.array([0,0,3,-2])

sum = vector_1 +vector_2
```

# Subtracting the vectors

```python
vector_1 = np.array([0,0,2,3])
vector_2= np.array([0,0,2,-3])

vector_subtraction = vector_1 - vector_2

print(vector_subtraction)
```
