# Projection Of a Vector Screenshot 2025-04-02 131742.png

```python
import numpy as np

a = np.array([10, 20])
v = np.array([23, -9])

magnitude_of_vector = np.sqrt(sum(v**2))

proj_a_on_v = (np.dot(a, v) / magnitude_of_vector**2) * v

print("Projection of a onto v:", proj_a_on_v)
```
